```python
import pandas as pd
import os
import requests
from bs4 import BeautifulSoup
import re

df=pd.read_csv('Pokemon3.csv',encoding='latin1')
pokemon_feature=df[['num','Name','HP','Attack','Defense','Speed','Generation','Legendary']].values.tolist()

#print(pokemon_feature)

#class:True or False 갯수 count
True_count=0
False_count=0
for i in range(len(pokemon_feature)):
    if (pokemon_feature[i][7]==True):
        True_count+=1
    else:
        False_count+=1
print("**********************True&False************************")
print("True_count : \t",True_count, "\tpercent : ", True_count/800)
print("False_count : \t",False_count, "\tpercent : ", False_count/800)

#low=1, middle=2, high=3
#HP count
T_HP_low=0
T_HP_middle=0
T_HP_high=0
F_HP_low=0
F_HP_middle=0
F_HP_high=0

for i in range(len(pokemon_feature)):
    if (pokemon_feature[i][7]==True):
        if (pokemon_feature[i][2]==1):
            T_HP_low+=1
        elif (pokemon_feature[i][2]==2):
            T_HP_middle+=1
        elif (pokemon_feature[i][2]==3):
            T_HP_high+=1
    else:
        if (pokemon_feature[i][2]==1):
            F_HP_low+=1
        elif (pokemon_feature[i][2]==2):
            F_HP_middle+=1
        elif (pokemon_feature[i][2]==3):
            F_HP_high+=1

print("**************************HP****************************")           
print("T_HP_low : \t", T_HP_low,"\tpercent : ",T_HP_low/True_count)
print("T_HP_middle : \t", T_HP_middle,"\tpercent : ",T_HP_middle/True_count)
print("T_HP_high : \t", T_HP_high,"\tpercent : ",T_HP_high/True_count)
print("F_HP_low : \t", F_HP_low,"\tpercent : ",F_HP_low/False_count)
print("F_HP_middle : \t", F_HP_middle,"\tpercent : ",F_HP_middle/False_count)
print("F_HP_high : \t", F_HP_high,"\tpercent : ",F_HP_high/False_count)


#low=1, middle=2, high=3
#Attack count
T_Attack_low=0
T_Attack_middle=0
T_Attack_high=0
F_Attack_low=0
F_Attack_middle=0
F_Attack_high=0

for i in range(len(pokemon_feature)):
    if (pokemon_feature[i][7]==True):
        if (pokemon_feature[i][3]==1):
            T_Attack_low+=1
        elif (pokemon_feature[i][3]==2):
            T_Attack_middle+=1
        elif (pokemon_feature[i][3]==3):
            T_Attack_high+=1
    else:
        if (pokemon_feature[i][3]==1):
            F_Attack_low+=1
        elif (pokemon_feature[i][3]==2):
            F_Attack_middle+=1
        elif (pokemon_feature[i][3]==3):
            F_Attack_high+=1

print("************************Attack**************************")           
print("T_Attack_low : \t\t", T_Attack_low,"\tpercent : ",T_Attack_low/True_count)
print("T_Attack_middle : \t", T_Attack_middle,"\tpercent : ",T_Attack_middle/True_count)
print("T_Attack_high : \t", T_Attack_high,"\tpercent : ",T_Attack_high/True_count)
print("F_Attack_low : \t\t", F_Attack_low,"\tpercent : ",F_Attack_low/False_count)
print("F_Attack_middle : \t", F_Attack_middle,"\tpercent : ",F_Attack_middle/False_count)
print("F_Attack_high : \t", F_Attack_high,"\tpercent : ",F_Attack_high/False_count)


#low=1, middle=2, high=3
#Defense count
T_Defense_low=0
T_Defense_middle=0
T_Defense_high=0
F_Defense_low=0
F_Defense_middle=0
F_Defense_high=0

for i in range(len(pokemon_feature)):
    if (pokemon_feature[i][7]==True):
        if (pokemon_feature[i][4]==1):
            T_Defense_low+=1
        elif (pokemon_feature[i][4]==2):
            T_Defense_middle+=1
        elif (pokemon_feature[i][4]==3):
            T_Defense_high+=1
    else:
        if (pokemon_feature[i][4]==1):
            F_Defense_low+=1
        elif (pokemon_feature[i][4]==2):
            F_Defense_middle+=1
        elif (pokemon_feature[i][4]==3):
            F_Defense_high+=1

print("************************Defense**************************")           
print("T_Defense_low : \t", T_Defense_low,"\tpercent : ",T_Defense_low/True_count)
print("T_Defense_middle : \t", T_Defense_middle,"\tpercent : ",T_Defense_middle/True_count)
print("T_Defense_high : \t", T_Defense_high,"\tpercent : ",T_Defense_high/True_count)
print("F_Defense_low : \t", F_Defense_low,"\tpercent : ",F_Defense_low/False_count)
print("F_Defense_middle : \t", F_Defense_middle,"\tpercent : ",F_Defense_middle/False_count)
print("F_Defense_high : \t", F_Defense_high,"\tpercent : ",F_Defense_high/False_count)



#low=1, middle=2, high=3
#Speed count
T_Speed_low=0
T_Speed_middle=0
T_Speed_high=0
F_Speed_low=0
F_Speed_middle=0
F_Speed_high=0

for i in range(len(pokemon_feature)):
    if (pokemon_feature[i][7]==True):
        if (pokemon_feature[i][5]==1):
            T_Speed_low+=1
        elif (pokemon_feature[i][5]==2):
            T_Speed_middle+=1
        elif (pokemon_feature[i][5]==3):
            T_Speed_high+=1
    else:
        if (pokemon_feature[i][5]==1):
            F_Speed_low+=1
        elif (pokemon_feature[i][5]==2):
            F_Speed_middle+=1
        elif (pokemon_feature[i][5]==3):
            F_Speed_high+=1

print("*************************Speed***************************")           
print("T_Speed_low : \t\t", T_Speed_low,"\tpercent : ",T_Speed_low/True_count)
print("T_Speed_middle : \t", T_Speed_middle,"\tpercent : ",T_Speed_middle/True_count)
print("T_Speed_high : \t\t", T_Speed_high,"\tpercent : ",T_Speed_high/True_count)
print("F_Speed_low : \t\t", F_Speed_low,"\tpercent : ",F_Speed_low/False_count)
print("F_Speed_middle : \t", F_Speed_middle,"\tpercent : ",F_Speed_middle/False_count)
print("F_Speed_high : \t\t", F_Speed_high,"\tpercent : ",F_Speed_high/False_count)



#Generation 1~6
#Generation count
T_Generation_1=0
T_Generation_2=0
T_Generation_3=0
T_Generation_4=0
T_Generation_5=0
T_Generation_6=0
F_Generation_1=0
F_Generation_2=0
F_Generation_3=0
F_Generation_4=0
F_Generation_5=0
F_Generation_6=0

for i in range(len(pokemon_feature)):
    if (pokemon_feature[i][7]==True):
        if (pokemon_feature[i][6]==1):
            T_Generation_1+=1
        elif (pokemon_feature[i][6]==2):
            T_Generation_2+=1
        elif (pokemon_feature[i][6]==3):
            T_Generation_3+=1
        elif (pokemon_feature[i][6]==4):
            F_Generation_4+=1
        elif (pokemon_feature[i][6]==5):
            F_Generation_5+=1
        elif (pokemon_feature[i][6]==6):
            F_Generation_6+=1
    else:
        if (pokemon_feature[i][6]==1):
            F_Generation_1+=1
        elif (pokemon_feature[i][6]==2):
            F_Generation_2+=1
        elif (pokemon_feature[i][6]==3):
            F_Generation_3+=1
        elif (pokemon_feature[i][6]==4):
            F_Generation_4+=1
        elif (pokemon_feature[i][6]==5):
            F_Generation_5+=1
        elif (pokemon_feature[i][6]==6):
            F_Generation_6+=1

print("*************************Generation***************************")           
print("T_Generation_1 : \t", T_Generation_1,"\tpercent : ",T_Generation_1/True_count)
print("T_Generation_2 : \t", T_Generation_2,"\tpercent : ",T_Generation_2/True_count)
print("T_Generation_3 : \t", T_Generation_3,"\tpercent : ",T_Generation_3/True_count)
print("T_Generation_4 : \t", T_Generation_4,"\tpercent : ",T_Generation_4/True_count)
print("T_Generation_5 : \t", T_Generation_5,"\tpercent : ",T_Generation_5/True_count)
print("T_Generation_6 : \t", T_Generation_6,"\tpercent : ",T_Generation_6/True_count)
print("F_Generation_1 : \t", F_Generation_1,"\tpercent : ",F_Generation_1/False_count)
print("F_Generation_2 : \t", F_Generation_2,"\tpercent : ",F_Generation_2/False_count)
print("F_Generation_3 : \t", F_Generation_3,"\tpercent : ",F_Generation_3/False_count)
print("F_Generation_4 : \t", F_Generation_4,"\tpercent : ",F_Generation_4/False_count)
print("F_Generation_5 : \t", F_Generation_5,"\tpercent : ",F_Generation_5/False_count)
print("F_Generation_6 : \t", F_Generation_6,"\tpercent : ",F_Generation_6/False_count)
```

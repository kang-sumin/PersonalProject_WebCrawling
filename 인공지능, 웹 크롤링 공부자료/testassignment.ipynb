```python
#enter_news뉴스에서 명사를 추출해서 단어의 빈도수를 뽑아내는 코드

import pandas as pd
df=pd.read_csv('enter_content.csv', encoding='utf-8-sig')
new_data=df['contents'].values.tolist()
all_data='\n'.join(new_data)


#명사추출
from konlpy.tag import Twitter
nlpy=Twitter()
nouns=nlpy.nouns(all_data)
#print(1,".예외처리 전",nouns)
stop_words=['위해','대해','우리','라며','때문','이번','마르고','대한','기사','뉴스','그때','죄송','작심','진짜','우려','불쑥','누가','동아일보','종부세','월세'] #예외하는 글자 리스트
nouns=[n for n in nouns if not n in stop_words and len(n)>1] #글자수가 한글자 이상이며 예외단어가 아닌것이 명사임

#print(2,".예외처리 후",nouns)

#단어의 빈도수 게산
from collections import Counter
count=Counter(nouns)
#tags=count.most_common(100)#명사 빈도수 상위 100개 출력
tags=count.most_common(200)

print(tags)
```

```python
df = pd.read_csv('enter_content.csv', encoding='utf-8-sig')  #파일 읽기
news_data=df['title'].values.tolist()

#명사 추출
from konlpy.tag import Twitter
all_data=[]
for n in news_data:
    nlpy = Twitter()
    nouns = nlpy.nouns(n)
    stop_words=['위해','대해','우리','라며','때문','오후']
    nouns = [n for n in nouns if not n in stop_words and len(n) > 1] #한글자수 빼기
    nouns=' '.join(nouns)
    all_data.append(nouns)


#TF-IDF 계산
from sklearn.feature_extraction.text import TfidfVectorizer
#TF-IDF vectorizer
tfv=TfidfVectorizer(ngram_range=(1,1))
#Transformation
vec_text=tfv.fit_transform(all_data)
print(vec_text)
```

```python
#Features
words=tfv.get_feature_names()
print(len(words))
print(words)
```

```python
#KMeans 계산                                                                           
from sklearn.cluster import KMeans
## K Means Clustering
kmeans = KMeans(n_clusters = 3, n_init = 17, n_jobs = -1, tol = 0.01, max_iter = 200)                                                                
#Fitting
kmeans.fit(vec_text)
## Transforming Numbers Back To Words.
print(kmeans.cluster_centers_.argsort())
print(kmeans.cluster_centers_)
#kmeans.labels_
```

```python
## Transforming Numbers Back To Words
common_words = kmeans.cluster_centers_.argsort()[:,-1:-11:-1]
for num, centroid in enumerate(common_words):
    print(str(num) + ' : ' + ', '.join(words[word] for word in centroid))
```

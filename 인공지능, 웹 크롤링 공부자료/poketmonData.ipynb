```python
#All OF Percent(변수에 percent정리)

#case T,F
Per_T=True_count/799
Per_F=False_count/799

#HP Percent
Per_T_HP_low=T_HP_low/True_count
Per_T_HP_middle=T_HP_middle/True_count
Per_T_HP_high=T_HP_high/True_count
Per_F_HP_low=F_HP_low/False_count
Per_F_HP_middle=F_HP_middle/False_count
Per_F_HP_high=F_HP_high/False_count

#Attack Percent
Per_T_Attack_low=T_Attack_low/True_count
Per_T_Attack_middle=T_Attack_middle/True_count
Per_T_Attack_high=T_Attack_high/True_count
Per_F_Attack_low=F_Attack_low/False_count
Per_F_Attack_middle=F_Attack_middle/False_count
Per_F_Attack_high=F_Attack_high/False_count

#Defense Percent
Per_T_Defense_low=T_Defense_low/True_count
Per_T_Defense_middle=T_Defense_middle/True_count
Per_T_Defense_high=T_Defense_high/True_count
Per_F_Defense_low=F_Defense_low/False_count
Per_F_Defense_middle=F_Defense_middle/False_count
Per_F_Defense_high=F_Defense_high/False_count

#Speed Percent
Per_T_Speed_low=T_Speed_low/True_count
Per_T_Speed_middle=T_Speed_middle/True_count
Per_T_Speed_high=T_Speed_high/True_count
Per_F_Speed_low=F_Speed_low/False_count
Per_F_Speed_middle=F_Speed_middle/False_count
Per_F_Speed_high=F_Speed_high/False_count

#Ceneration Percent
Per_T_Generation_1=T_Generation_1/True_count
Per_T_Generation_2=T_Generation_2/True_count
Per_T_Generation_3=T_Generation_3/True_count
Per_T_Generation_4=T_Generation_4/True_count
Per_T_Generation_5=T_Generation_5/True_count
Per_T_Generation_6=T_Generation_6/True_count
Per_F_Generation_1=F_Generation_1/False_count
Per_F_Generation_2=F_Generation_2/False_count
Per_F_Generation_3=F_Generation_3/False_count
Per_F_Generation_4=F_Generation_4/False_count
Per_F_Generation_5=F_Generation_5/False_count
Per_F_Generation_6=F_Generation_6/False_count
```

```python
#case 실행
from random import*


case=False
count=1
while(case!=True):
    #create Random case number
    R_HP=randint(0,200)
    R_Attack=randint(0,200)
    R_Defense=randint(0,200)
    R_Speed=randint(0,200)
    R_Generation=randint(1,6)
    
    #class case 변수
    case_T=1
    case_F=1

    #case HP
    if (R_HP<=85):
        case_T*=Per_T_HP_low
        case_F*=Per_F_HP_low
    elif (R_HP>85 and R_HP<=170):
        case_T*=Per_T_HP_middle
        case_F*=Per_F_HP_middle
    elif (R_HP>170):
        case_T*=Per_T_HP_high
        case_F*=Per_F_HP_high
    #print(i+1,'\tHP\t',case_T, "\tvs\t",case_F)
    print(count,'\t:\t',"T:",case_T, "\tvs\t","F:",case_F)

    #case Attack
    if (R_Attack<=63):
        case_T*=Per_T_Attack_low
        case_F*=Per_F_Attack_low
    elif (R_Attack>63 and R_Attack<=126):
        case_T*=Per_T_Attack_middle
        case_F*=Per_F_Attack_middle
    elif (R_Attack>126):
        case_T*=Per_T_Attack_high
        case_F*=Per_F_Attack_high
    #print(i+1,'\tAttack\t',case_T, "\tvs\t",case_F)
    print(count,'\t:\t',"T:",case_T, "\tvs\t","F:",case_F)

    #case Defense
    if (R_Defense<=76):
        case_T*=Per_T_Defense_low
        case_F*=Per_F_Defense_low
    elif (R_Defense>76 and R_Defense<=152):
        case_T*=Per_T_Defense_middle
        case_F*=Per_F_Defense_middle
    elif (R_Defense>152):
        case_T*=Per_T_Defense_high
        case_F*=Per_F_Defense_high
    print(count,'\t:\t',"T:",case_T, "\tvs\t","F:",case_F)

    #case Speed
    if (R_Speed<=60):
        case_T*=Per_T_Speed_low
        case_F*=Per_F_Speed_low
    elif (R_Speed>60 and R_Speed<=152):
        case_T*=Per_T_Speed_middle
        case_F*=Per_F_Speed_middle
    elif (R_Speed>152):
        case_T*=Per_T_Speed_high
        case_F*=Per_F_Speed_high
    print(count,'\t:\t',"T:",case_T, "\tvs\t","F:",case_F)

    #case Generation
    if (R_Generation==1):
        case_T*=Per_T_Generation_1
        case_F*=Per_F_Generation_1
    elif (R_Generation==2):
        case_T*=Per_T_Generation_2
        case_F*=Per_F_Generation_2
    elif (R_Generation==3):
        case_T*=Per_T_Generation_3
        case_F*=Per_F_Generation_3
    elif (R_Generation==4):
        case_T*=Per_T_Generation_4
        case_F*=Per_F_Generation_4
    elif (R_Generation==5):
        case_T*=Per_T_Generation_5
        case_F*=Per_F_Generation_5
    elif (R_Generation==6):
        case_T*=Per_T_Generation_6
        case_F*=Per_F_Generation_6
    print(count,'\t:\t',"T:",case_T, "\tvs\t","F:",case_F)
    
    #case T,F multiply
    case_T*=Per_T
    case_F*=Per_F
    #print(count,'\t:\t',"T:",case_T, "\tvs\t","F:",case_F)
    
    if(case_T>case_F):
        case=True
        print(count,'\t',"T:",case_T, "\tvs\t","F:",case_F,count,"번째 포켓몬은 Legend Pokemon 입니다!!!")
        #print(count,"번째 포켓몬은 Legend Pokemon 입니다!!!")
    else:
        case=False
        print(count,'\t',"T:",case_T, "\tvs\t","F:",case_F,count,"번째 포켓몬은 Legend Pokemon 아닙니다...")
        #print(count,"번째 포켓몬은 Legend Pokemon 아닙니다... ")
    
    count+=1
    
count-=1
print("**********\t",count,"번만에 Legend Pokemon이 나왔습니다!!!\t**********")
    
```

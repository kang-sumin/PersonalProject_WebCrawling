```python
import requests
from bs4 import BeautifulSoup
import pandas as pd   #pandas라이브러리를 사용 할때 ps로 줄여서 사용
import os
df = pd.read_csv('new_group_lists1.csv', encoding='cp949')  #파일 읽기
news_data=df[['group','labels']].values.tolist()
link_group=[]
for group, labels in news_data:
    if labels==1:
        webpage = requests.get(group)
        soup = BeautifulSoup(webpage.content, "html.parser", from_encoding='utf-8')
        link_data=soup.find_all('div', {'class': 'rightList'} )
        for tag in link_data:
             link_group.append([tag.find('a')['href'],labels])
    else:
        webpage = requests.get(group)
        soup = BeautifulSoup(webpage.content, "html.parser", from_encoding='utf-8')
        link_data = soup.find_all('dl',{'class': 'list_item'})
        for tag in link_data:
             link_group.append(['https:'+tag.find('a')['href'],labels])

df = pd.DataFrame(link_group, columns=['url', 'labels'])   #DataFrame을 생성 데이터값과 컬럼 생성   
df.to_csv('news_url.csv', index=False, mode='w', encoding='cp949')  #파일 생성 mode='w' 한글 저장 encoding='cp949' 
```

```python
import pandas as pd   #pandas라이브러리를 사용 할때 ps로 줄여서 사용
import os
import requests
from bs4 import BeautifulSoup
df = pd.read_csv('news_url.csv', encoding='cp949')  #파일 읽기
news_url=df[['url','labels']].values.tolist()
news_data=[]
for url, labels in news_url:
    if labels==1:
        webpage = requests.get(url)
        soup = BeautifulSoup(webpage.content, "html.parser", from_encoding='utf-8')
        title=soup.find('h1', {'class': 'title'} )
        data=soup.find('div', {'class': 'article_txt'} )
        news_data.append([title.text,data.text])
    else:
        webpage = requests.get(url)
        soup = BeautifulSoup(webpage.content, "html.parser", from_encoding='utf-8')
        title=soup.find('h1', {'id': 'news_title_text_id'} )
        data=soup.find('div', {'id': 'news_body_id'} )
        news_data.append([title.text,data.text])

df = pd.DataFrame(news_data, columns=['title', 'news'])   #DataFrame을 생성 데이터값과 컬럼 생성   
df.to_csv('news_data1.csv', index=False, mode='w', encoding='utf-8-sig')  #파일 생성 mode='w' 한글 저장 encoding='cp949' 
```
